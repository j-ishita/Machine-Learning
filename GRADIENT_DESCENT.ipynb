```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#function

def fx(x):
  return (3*(x**2))-(3*x)+4
```

```python
#derivative function
def derv(x):
  return (6*x)-3
```

```python
# define a range for x
x = np.linspace(-2,2,2001)
```

```python
#plot
plt.plot(x,fx(x),x, derv(x))
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y', 'dy'])
```

```python
#random starting point
lmin = np.random.choice(x,1)
print(lmin)

#learning rate and number of training cycle

learning_rate = 0.01
epochs1=100

#run iteration through training cycles
for i in range(epochs1):
  grad = derv(lmin)
  lmin = lmin-(learning_rate*grad)

lmin

#run through cycles and store the results
params = np.zeros((epochs1,2))

for i in range(epochs1):
  grad = derv(lmin)
  lmin = lmin-(learning_rate*grad)
  params[i,0] = lmin
  params[i,1] = grad
```

```python
#plot the gradient over iterations

fig,ax=plt.subplots(1,2, figsize=(12,5))

for i in range(2):
  ax[i].plot(params[:,i])
  ax[i].set_xlabel('iteration')
ax[0].set_ylabel('Local min')
ax[1].set_ylabel('Local min')
```
